#### Plotting Different Graphs by matplotlib and seaborn.

```python
plt.plot(df["brand"])
plt.ylabel("Brand Names")
plt.show()
```

```python
sns.lineplot(df["price"])
plt.show()
```

```python
plt.hist(df["owner"])
plt.show()
```

```python
sns.histplot(x="power",data=df)
plt.title("Power Distribution Graph")
plt.show()
```

```python
sns.countplot(x="age",data=df)
plt.title("Age count graph")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.kdeplot(x="power",data=df)
plt.show()
```

```python
plt.scatter(df["age"],df["price"])
plt.show()
```

```python
sns.scatterplot(x="power",y="price",hue="owner",data=df)
plt.show()
```

```python
sns.histplot(x = "brand",data=df,kde=True)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(y="kms_driven",hue="owner",kind="bar",data=df)
plt.title("kms_driven vs owner")
plt.show()
```

```python
owner = df["owner"].value_counts()
plt.pie(owner,labels=owner.index,autopct='%1.1f%%',explode=[0,0,0,0.2],shadow=True)
plt.title("Owner percentage")
plt.show()
```

```python
sns.relplot(y="kms_driven",x="price",hue="owner",data=df)
plt.show()
```

```python
predict = df.groupby("owner")[[
    "age",
    "power",
    "price",
    "kms_driven"
]].mean()

predict.plot(kind="bar")
plt.title("Saling of bike")
plt.ylabel("factors")
plt.xticks(rotation=60)
plt.show()
```
